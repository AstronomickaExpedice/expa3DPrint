```python
from solid import *
from solid.utils import *
```

```python
cl = 0.1
M5 = {'d':5+0.4, 'k': 1, 'dk': 5.5+0.5}
M3 = {'d':3+0.4, 'k': 1, 'dk': 8.5+0.5}
podlozka1 = {'d': 5, 'D': 10, 'B': 0.5}
bearing_625 = {'d': 5, 'D': 16, 'B': 5}
```

```python
az_bearing = bearing_625
wall = 3
```

```python
def boltHole(bolt, l = 12, hex_head = True, hex_nut = True):
    p = cylinder(h=l, d=bolt['d'])
    p+= down(l)(
            if hex_head:
                cylinder(d=5, h=l, segments=6)
            else:
                cylinder(d=5, h=l)
        )
    return p
```

```python
#az_1 = cylinder(d=az_bearing['D'] + 2*wall, h = az_bearing['B']+wall)
az_1 = cylinder(d1=30, d2=az_bearing['D'] + 2*wall, h=az_bearing['B']+wall)
az_1-= cylinder(d=az_bearing['D'], h=az_bearing['B'])
az_1-= cylinder(d=M5['d'], h=az_bearing['B']+wall)
for x in range(4):
    az_1 -= rotate([0,0, x*90])(
        translate([0, 10, 0])(
            cylinder(d=M3['d'], h=100),
            up(az_bearing['B']+wall)(
                rotate([180, 0, 0])(
                    
                    debug(boltHole(M3, l=10))
                )
            )
        )
    )
    
    
    scad_render_to_file(az_1, '../az_1.scad', file_header='$fn = 100;', include_orig_code=True)
```
